```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('funda.csv', sep=';')
df.head()
```

```python
df_subset = df[df['price'] > 100000]
print(df_subset.head())
```

```python
sns.pairplot(df_subset)
plt.show() #using Matplotlib to show the plot
```

```python
sns.lmplot(x='bedrooms', y='rooms', fit_reg=False, data=df) #fit_reg=False leaves out the regression line; see below
plt.xlabel('Subscribers (millions)') #use matplotlib to refine Seaborn plot
plt.ylabel('Views (millions)')
plt.title('YouTube channels subscribers and views')
plt.show()
```

```python
corr = df[['bedrooms', 'price']].corr() #I'm leaving out the corrected variables (divided by 1 million)
corr
```
