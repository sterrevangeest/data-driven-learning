```python
import pandas as pd
```

```python
df_steps = pd.read_csv('steps.csv', sep=';')
df_steps.head()
```

```python
df_survey = pd.read_csv('survey.csv')
df_survey.head()
```

```python
df = pd.merge(df_survey, df_steps, on = 'id') #'app' is the common identifier
```

```python
df.head()
```

```python
# print(df["height"],df["weight"].sort_values())
```

```python
def above_5(x):
    if(x > 155.0 or x < 45.0 ): 
        return float('NaN')
    else: 
        return x

df['weight'] = df['weight'].apply(above_5)
df["weight"].sort_values()
```
